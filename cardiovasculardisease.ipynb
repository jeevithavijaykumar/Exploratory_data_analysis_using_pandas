```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cvddf = pd.read_csv(r'./sample_data/CVD_cleaned.csv')
```

```python
cvddf.head()
```

```python
cvddf.info()
```

```python
cvddf.shape
```

```python
#Drop rows with missing values
cvddf.dropna()
```

```python
#Check for duplicate rows
cvddf.duplicated().value_counts()
```

```python
#Drop duplicate rows
cvddf.drop_duplicates()
```

```python
#Distribution of General health among patients
import seaborn as sns
x= cvddf['General_Health'].value_counts().index
y=cvddf['General_Health'].value_counts().values
plt.figure(figsize=(7,4))
plt.bar(x,y, color='lightblue')
plt.title('General Health Distribution Among Patients')
plt.xlabel('General helath')
plt.ylabel('Count')
plt.legend
plt.show()
```

```python
#Distribution of checkup frequency among patients
x1= cvddf['Checkup'].value_counts().index
y1= cvddf['Checkup'].value_counts().values
plt.figure(figsize=(8,4))
sns.barplot(x=y1,y=x1,palette='pastel')
plt.title('Distribution of checkup frequency among patients')
plt.xlabel('No of patients')
plt.ylabel('checkup frequency')
plt.legend
plt.show()
```

```python
#Excercise participation rate
plt.figure(figsize=(4,4))
x= cvddf['Exercise'].value_counts().index
y=cvddf['Exercise'].value_counts().values
plt.pie(y, labels=x, colors=sns.color_palette('pastel'), autopct='%1.1f%%', explode=[0,0.02])
plt.title('Excercise participation rate among patients')
plt.legend
plt.show()
```

```python
#Correlation Between Exercise and Heart Disease
plt.figure(figsize=(7,4))
sns.countplot(data=cvddf,x= cvddf['Heart_Disease'], hue=cvddf['Exercise'], palette='pastel')
plt.show()
```

```python
#Arthritis Prevalence Among Patients by Age Group
plt.figure(figsize=(13,4))
sns.countplot(data=cvddf, x=cvddf['Age_Category'], hue=cvddf['Arthritis'], palette='pastel')
plt.title('Arthritis Prevalence Among Patients by Age Group')
plt.legend
plt.show()
```

```python
#percentage of patients diagnosed with skin cancer.
plt.figure(figsize=(4,4))
x= cvddf['Skin_Cancer'].value_counts().index
y=cvddf['Skin_Cancer'].value_counts().values
plt.pie(y, labels=x, colors=sns.color_palette('pastel'), autopct='%1.1f%%', explode=[0,0.02])
plt.title('Percentage of Patients diagnosed with skin cancer')
plt.legend
plt.show()
```

```python
#Prevalence of depression among patients
plt.figure(figsize=(7,4))
sns.countplot(data=cvddf, x=cvddf['Sex'], hue=cvddf['Depression'], palette='pastel')
plt.title('Prevalence of depression among patients')
plt.legend
plt.show()
```

```python
#Diabetic Status Among Patients
cvddf['Diabetes'].value_counts()
```

```python
#Smoking History of All Patients
plt.figure(figsize=(4,4))
x= cvddf['Smoking_History'].value_counts().index
y=cvddf['Smoking_History'].value_counts().values
plt.pie(y, labels=x, colors=sns.color_palette('pastel'), autopct='%1.1f%%', explode=[0,0.01])
plt.title('Smoking History of Patients')
plt.legend
plt.show()
```

```python
#Histogram
plt.figure(figsize=(18,7))
plt.subplot(2,3,1)
cvddf['BMI'].hist(bins=20,color='lightblue')
plt.xlabel('BMI')
plt.subplot(2,3,2)
cvddf['Alcohol_Consumption'].hist(bins=15,color='lightblue')
plt.xlabel('Alcohol Consumption')
plt.subplot(2,3,3)
cvddf['Fruit_Consumption'].hist(bins=15,color='lightblue')
plt.xlabel('Fruit Consumption')
plt.subplot(2,3,4)
cvddf['Green_Vegetables_Consumption'].hist(bins=15,color='lightblue')
plt.xlabel('Green Vegetables Consumption')
plt.subplot(2,3,5)
cvddf['FriedPotato_Consumption'].hist(bins=15,color='lightblue')
plt.xlabel('Fried Potato Consumption')
plt.legend
plt.subplots_adjust(wspace=0.25, hspace=0.5)
plt.show()
```
