# *Exploratory Data Analysis*

```python
#Sales Channel Distribution Across Total Bookings
x=bookingdf['sales_channel'].value_counts().index
y=bookingdf['sales_channel'].value_counts().values
plt.figure(figsize=(2.5,4))
plt.title('Sales Channel Distribution Across Total Bookings')
plt.bar(x,y,color='teal')
plt.xlabel('sale channel')
plt.ylabel('No of bookings')
plt.show()
plt.legend
```

```python
#Distribution Of Trip Types In Total Booking'
x= bookingdf['trip_type'].value_counts().index
y=bookingdf['trip_type'].value_counts().values
plt.figure(figsize=(4,4))
plt.title('Distribution Of Trip Types In Total Bookings')
plt.bar(x,y,color='plum')
plt.xlabel('Trip type')
plt.ylabel('No. of bookings')
plt.show()
plt.legend
```

```python
#Histogram for Numerical variables
col=['num_passengers','length_of_stay','purchase_lead','flight_hour']
bookingdf[col].hist(bins=20, color ='teal',figsize=(15,6))
plt.show()
```

```python
#Distribution of bookings over the days of the week
import seaborn as sns
x1 = bookingdf['flight_day'].value_counts().index
y1= bookingdf['flight_day'].value_counts().values
plt.figure(figsize=(8,4))
sns.barplot(x=y1,y=x1,palette='crest')
plt.xlabel('Number of bookings')
plt.ylabel('days of the week')
plt.title('Booking Distribution Across The Days Of The Week')
plt.show()
plt.legend
```

```python
#Top ten most popular routes
x= bookingdf['route'].value_counts().nlargest(10).index
y= bookingdf['route'].value_counts().nlargest(10).values
c =['steelblue','lightgreen','rosybrown','darkkhaki','slateblue','teal','lavender','orchid','tan','darkslategray']
plt.figure(figsize=(5,5))
plt.pie(y, labels=x, autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Top Ten Most Popular Routes')
plt.show()
plt.legend
```

```python
#Service Required vs Booking completion
#services = ['wants_extra_baggage','wants_preferred_seat','wants_in_flight_meals']
p=['teal','plum']
plt.figure(figsize=(15,3.5))
plt.subplot(1,3,1)
sns.countplot(bookingdf, x=bookingdf['wants_extra_baggage'], hue='booking_complete', palette=p)
plt.subplot(1,3,2)
sns.countplot(bookingdf, x=bookingdf['wants_preferred_seat'], hue='booking_complete', palette=p)
plt.subplot(1,3,3)
sns.countplot(bookingdf, x=bookingdf['wants_in_flight_meals'], hue='booking_complete', palette=p)
plt.subplots_adjust(wspace=0.5)
plt.show()
```

```python
#Percentage of Completed Bookings among Total Bookings
plt.figure(figsize=(4,4))
x= bookingdf['booking_complete'].value_counts().index
y= bookingdf['booking_complete'].value_counts().values
plt.pie(y, labels=x, autopct='%1.1f%%', colors=p, explode=[0,0.1])
plt.title('Percentage of Completed Bookings among Total Bookings')
plt.show()
```

```python
#Distribution of Purchase lead
plt.figure(figsize=(7,4))
x1=bookingdf['purchase_lead']. value_counts().nlargest(10).index
y1=bookingdf['purchase_lead']. value_counts().nlargest(10).values
sns.barplot(x=x1,y=y1,palette='crest')
plt.ylabel('Purchase lead')
plt.show()
plt.legend
```
