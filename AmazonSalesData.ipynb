```python
salesdf['discounted_price'].str.contains('₹').sum()
salesdf['discounted_price'].str.contains(',').sum()
salesdf['discounted_price']= salesdf['discounted_price'].str.replace('₹','').str.replace(',','')
```

```python
salesdf['actual_price'].str.contains('₹').sum()
salesdf['actual_price'].str.contains(',').sum()
salesdf['actual_price'] = salesdf['actual_price'].str.replace('₹','').str.replace(',','')
```

```python
salesdf['discount_percentage'].str.contains('%').sum()
salesdf['discount_percentage'] = salesdf['discount_percentage'].str.replace('%','')
```

```python
salesdf['rating_count'].str.contains(',').sum()
salesdf['rating_count']=salesdf['rating_count'].str.replace(',','')
```

```python
salesdf['discounted_price'] =salesdf['discounted_price'].astype('float64')
salesdf['actual_price']=salesdf['actual_price'].astype('float64')
salesdf['discount_percentage']=salesdf['discount_percentage'].astype('float64')
```

```python
salesdf['rating']=salesdf['rating'].astype('float64')
```

```python
#remove rows with ratings '|'
salesdf= salesdf.loc[salesdf['rating'] != '|']
```

```python
salesdf['rating']=salesdf['rating'].astype('float64')
salesdf['rating_count']=salesdf['rating_count'].astype('float64')
```

```python
#Split Categories
salesdf[['Product_Division','Product_Group','Product_Subgroup','Product_type',
        'Category_4','Category_5','Category_6']]= salesdf['category'].str.split('|', expand=True)
```

```python
salesdf.head()
```

```python
# Retrieve necessary columns for data visualization and further analysis to explore the relationships between variables.
finaldata = salesdf[['product_id','product_name','Product_Division','Product_Group','Product_Subgroup','Product_type',
                     'discounted_price','actual_price','discount_percentage','rating','rating_count']]
```

```python
#Top Five Product Divisions with the Highest Number of Products Sold
import seaborn as sns
x1 =finaldata['Product_Division'].value_counts().nlargest(5).index
y1 = finaldata['Product_Division'].value_counts().nlargest(5).values
plt.figure(figsize=(7,3))
sns.barplot(x=y1, y=x1, palette='viridis')
plt.title('Top Five Product Divisions with the Highest Number of Products Sold')
plt.legend
plt.show()
```

```python
#Variation of Discounts across Product Divisions
x1 = finaldata['Product_Division']
y1=finaldata['discount_percentage']
plt.figure(figsize=(16,2.5))
plt.bar(x1,y1,color=sns.color_palette('Blues'))
plt.title('Variation of Discounts across Product Divisions')
plt.xlabel('Product Divisions')
plt.ylabel('Discounts in Percentage')
plt.legend
plt.show()
```

```python
#Top Ten Product Types with the Highest Discounts Among All Discounted Products
x=finaldata['Product_type'].value_counts().nlargest(10).index
y=finaldata['Product_type'].value_counts().nlargest(10).values
plt.figure(figsize=(5,5))
plt.pie(y, labels=x, colors= sns.color_palette('Blues'), autopct='%1.1f%%')
plt.title('Top Ten Product Types Offering the Maximum Number of Discounted Products')
plt.legend
plt.show()
```

```python
#Top Ten Higly Rated Products
ratingdf = finaldata.sort_values('rating',ascending=False).nlargest(10,'rating')
y1=ratingdf['product_id']
x1=ratingdf['rating']
plt.figure(figsize=(6,3.5))
sns.barplot(x=x1,y=y1, palette='viridis')
plt.title('Top Ten Higly Rated Products')
plt.xlabel('Ratings')
plt.ylabel('Product ID')
plt.legend
plt.show()
```

```python
#Plot distribution of Actual And Discounted Price
plt.figure(figsize=(15,3))
plt.subplot(1,2,1)
finaldata['actual_price'].hist(bins=40,color='seagreen')
plt.xlabel('Actual price')
plt.ylabel('Frequency')
plt.title('Distribution of Actual Price')
plt.subplot(1,2,2)
plt.xlabel('Discounted price')
plt.ylabel('Frequency')
finaldata['discounted_price'].hist(bins=40,color='seagreen')
plt.title('Distribution of Discounted Price')
plt.subplots_adjust(wspace=0.25)
plt.show()
```
